## Lab 3: Exploratory Data Analysis (EDA)

In this lab, we perform exploratory data analysis on the Steam Games dataset.


```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look cleaner
sns.set()
```

```python
# Load Dataset
df = pd.read_csv("games.csv")

# Display first 5 rows
df.head()
```

### Check Missing Values
We check if there are missing values in the dataset.


```python
df.isna().sum()
```

```python
# checking duplicate rows
df.duplicated().sum()
```

### Number of Rows and Columns


```python
print("Shape:", df.shape)
```

```python
# viewing the data types of columns
df.dtypes
```

```python
# Statistical summary
df.describe()
```

## Univariate Analysis
### Distribution of Metacritic Scores


```python
plt.figure(figsize=(8,5))
sns.histplot(df['Metacritic score'].dropna(), bins=20)
plt.title("Distribution of Metacritic Scores")
plt.show()
```

Most games have mid-range ratings.


## Bivariate Analysis
### Price vs Metacritic Score


```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Price', y='Metacritic score', data=df)
plt.title("Price vs Metacritic Score")
plt.show()
```

There is no strong relationship between price and rating.


### Positive vs Negative Reviews


```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Positive', y='Negative', data=df)
plt.title("Positive vs Negative Reviews")
plt.show()
```

Games with more positive reviews also tend to have more negative reviews.

### Correlation Matrix

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['Price','Metacritic score','Positive','Negative']].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

The correlation matrix shows relationships between numerical features.
